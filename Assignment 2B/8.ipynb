```python
import pandas as pd
import re
```

```python
data = pd.read_csv('tittrain.csv')
old = pd.read_csv('7.csv')
```

```python
data.head()
```

```python
data['Title'] = old['Title']
data.head()
```

```python
counts = []
means = []
for tit in data['Title']:
    counts.append(data.loc[(data['Title']==tit)].shape[0])
    means.append(data.loc[(data['Title']==tit) & (data['Survived']==1)].shape[0] / data.loc[(data['Title']==tit)].shape[0] * 100)

    
    
    
```

```python
frame = pd.DataFrame({'Name':old['Name'], 'Title':old['Title'], 'Title_count':counts,'Title_mean':means})
frame = frame[['Name','Title','Title_count','Title_mean']]
sel = frame
sel.head()
```

```python
sel.to_csv("8.csv", index=False)
```
